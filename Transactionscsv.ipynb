```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("transactions.csv")
df
```

```python
# what is the  total sales amount across all transactions?
total_sales = df["t_amt"].sum()
print(total_sales)
```

```python
df['t_date'] = pd.to_datetime(df['t_date'])
df['Month'] = df['t_date'].dt.month
df['MonthName'] = df['t_date'].dt.strftime('%B')
df['Quarter'] = df['t_date'].dt.quarter
```

```python
# Find the month with the highest total transaction amount
c=df.groupby('MonthName')['t_amt'].sum()
print(c.idxmax())
```

```python
# Calculate average transaction amount per customer
d=df.groupby('cust_id')['t_amt'].mean()
print(d)
```

```python
# trend of total sales over months
trend= df.groupby("Month")["t_amt"].sum()
print(trend)
```

```python
#Find the highest single transaction amount
df['t_amt'].max()
```

```python
#Service category contributing most revenue
a=df.groupby('services')['t_amt'].sum()
print(a.idxmax())
```

```python
#Product generating highest revenue
b=df.groupby('products_used')['t_amt'].sum()
print(b.idxmax())
```

```python
#average transaction amount between services
c=df.groupby('services')['t_amt'].mean().sort_values(ascending=False)
print(c)
```

```python
#Number of unique customers
df['cust_id'].nunique()
```

```python
#Customers who spent most overall
df.groupby('cust_id')['t_amt'].sum().sort_values(ascending=False)
```

```python
# Average number of transactions per customer
df.groupby('cust_id').size().mean()
```

```python
#Customers who purchased from multiple services
df.groupby('cust_id')['services'].nunique().sort_values(ascending=False)
```

```python
#Percentage of repeat buyers
counts = df.groupby('cust_id').size()
((counts > 1).sum() / counts.count()) * 100
```

```python
#Service with highest total sales
df.groupby('services')['t_amt'].sum().sort_values(ascending=False)
```

```python
#Most popular services by transaction count
df['services'].value_counts()
```

```python
#Most frequently purchased product per service
df.groupby(['services','products_used']).size().sort_values(ascending=False)
```

```python
# Average transaction amount per product
df.groupby('products_used')['t_amt'].mean()
```

```python
# Services where customers spend more on average
df.groupby('services')['t_amt'].mean().sort_values(ascending=False)
```

```python
#State with highest total sales
df.groupby('state')['t_amt'].sum().sort_values(ascending=False)
```

```python
#City with highest number of transactions
df['city'].value_counts()
```
